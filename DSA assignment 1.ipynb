```python
import numpy as np

data = np.arange(3*16*10).reshape(3,16,10)

print("Original Data")
print(data)

```

```python
# Year-wise average
year_avg = np.mean(data, axis=(1,2))
print("Year-wise Average")
print(year_avg)
```

```python
# Student-wise total
student_total = np.sum(data, axis=2)
print("Student-wise Total")
print(student_total)
```

```python
# Subject-wise highest score
subject_max = np.max(data, axis=(0,1))
print("Subject-wise Highest Score")
print(subject_max)
```

```python
# Replace marks below 40 with 40
data[data < 40] = 40
print("Updated Data")
print(data)
```

```python
# Sort marks subject-wise
sorted_data = np.sort(data, axis=2)
print("Sorted Marks Subject-wise")
print(sorted_data)
```
