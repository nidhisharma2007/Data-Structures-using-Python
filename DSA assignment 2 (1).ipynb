# Calculators

```python
#Calculator
def calculator(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        return a / b
    elif op == "**":   
        return a ** b

a=7
b=8
op="+"

final=calculator(a,b,op)
print(final)
```

```python
#Discount calculator
def discount(price, percent):
    final_price = price - (price * percent / 100)
    return final_price

price = float(input("Enter the original price: "))
percent = float(input("Enter the discount percentage: "))

result = discount(price, percent)

print("Final price after discount:", result)
```

```python
#Grade calculator
def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "Fail"

marks = float(input("Enter your marks: "))

result = grade(marks)

print("Your grade is:", result)
```

```python
#Even odd checker
def even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

n = int(input("Enter a number: "))

result = even_odd(n)

print("The number is:", result)
```

```python
# Quadratic equation calculator
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b*b - 4*a*c  

if d > 0:
    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)
    print("Root 1:", x1)
    print("Root 2:", x2)

elif d == 0:
    x = -b / (2*a)
    print("Both roots are equal:", x)

else:
    print("Roots are imaginary (complex)")
```

# Sorting

```python
#bubble sort
def BBS(array):
  n = len(array)
  for i in range(n-1):
    for j in range (n-1-i):
      if array[j] > array[j+1]:
        array[j], array[j+1] = array[j+1],array[j]
  return array

data= [5, 4, 3, 2, 1]
print(BBS(data))
```

```python
#Selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
              min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
arr = [64, 25, 12, 22, 11]
print(selection_sort(arr))
```

```python
#insertion sort
arr = [5, 4, 3, 2, 1]
for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
print(arr)
```

```python
# Quick sort
def quick_sort(arr):
    if len(arr) <= 1: # Base Case
        return arr
    pivot = arr[-1]   
    left = []
    right = []
    for i in range(len(arr) - 1):
        if arr[i] < pivot:
            left.append(arr[i])
        else:
            right.append(arr[i])
    return quick_sort(left) + [pivot] + quick_sort(right)

arr = [8, 3, 1, 7, 0, 10, 2]
sorted_arr = quick_sort(arr)
print("Sorted array:", sorted_arr)
```
